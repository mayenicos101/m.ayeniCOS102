```python
#With libraries
import numpy as np
def quadratic_roots(A, B, C):
    coefficients = [A, B, C]
    roots = np.roots(coefficients)
    return roots
def cubic_roots(A, B, C, D):
    coefficients = [A, B, C, D]
    roots = np.roots(coefficients)
    return roots
def quartic_roots(A, B, C, D, E):
    coefficients = [A, B, C, D, E]
    roots = np.roots(coefficients)
    return roots
while True:    
    print("Select the type of polynomial:")
    print("1. Quadratic (ax^2 + bx + c = 0)")
    print("2. Cubic (ax^3 + bx^2 + cx + d = 0)")
    print("3. Quartic (ax^4 + bx^3 + cx^2 + dx + e = 0)")
    print("4. Exit")
    choice = input("Enter your choice (1-4): ")
    
    if choice == '1':
        A = float(input("Enter coefficient a: "))
        B = float(input("Enter coefficient b: "))
        C = float(input("Enter coefficient c: "))
        roots = quadratic_roots(A, B, C)
        print("The roots of the quadratic equation are:", roots)
    elif choice == '2':
        A = float(input("Enter coefficient a: "))
        B = float(input("Enter coefficient b: "))
        C = float(input("Enter coefficient c: "))
        D = float(input("Enter coefficient d: "))
        roots = cubic_roots(A, B, C, D)
        print("The roots of the cubic equation are:", roots)
    elif choice == '3':
        A = float(input("Enter coefficient a: "))
        B = float(input("Enter coefficient b: "))
        C = float(input("Enter coefficient c: "))
        D = float(input("Enter coefficient d: "))
        E = float(input("Enter coefficient e: "))
        roots = quartic_roots(A, B, C, D, E)
        print("The roots of the quartic equation are:", roots)
    elif choice == '4':
        print("Exiting...")
        break
    else:
        print("Invalid choice. Please enter a number between 1 and 4.")
```
