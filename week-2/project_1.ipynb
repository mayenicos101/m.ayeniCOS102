```
START
SET total = 0
SET counter = 1
DISPLAY "Enter number of items:"
INPUT itemCount
LOOP
    DISPLAY "Enter price of item:"
    INPUT price
    total = total + price
    counter = counter + 1
    IF counter > itemCount THEN
        EXIT LOOP
    END IF
END LOOP
DISPLAY "Total price is: ", total
DISPLAY "Enter payment amount"
INPUT payment
IF payment < total
    DISPLAY "Insufficient payment. Please enter a valid amount."
INPUT extraPayment
    payment = payment + extraPayment
END IF
CHANGE = payment - total
DISPLAY "Change = ", CHANGE
DISPLAY "Thank you for your purchase"
```
