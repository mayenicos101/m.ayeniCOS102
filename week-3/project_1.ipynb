```python
import tkinter as tk
from tkinter import messagebox
import csv

# Load employee data from CSV
employees = []
with open('GIG-logistics.csv', 'r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        employees.append(row)

def check_employee():
    name = name_entry.get().strip()
    dept = dept_entry.get().strip()
    
    # Find the employee
    employee_found = None
    for emp in employees:
        if emp['Name'].lower() == name.lower() and emp['Department'].lower() == dept.lower():
            employee_found = emp
            break
    
    if employee_found:
        # Welcome message
        message = f"Welcome {employee_found['Name']} from {employee_found['Department']}!\n\n"
        
        # List other members in the department
        others = [emp['Name'] for emp in employees if emp['Department'] == dept and emp['Name'] != name]
        if others:
            message += f"Other members in {dept}: {', '.join(others)}"
        else:
            message += f"You are the only member in {dept}."
        
        result_label.config(text=message, fg="green")
    else:
        result_label.config(text="Sorry, the employee does not exist in our records.", fg="red")

# Create GUI
root = tk.Tk()
root.title("GIG Logistics Employee Verification")
root.geometry("400x300")

tk.Label(root, text="Employee Name:").pack(pady=5)
name_entry = tk.Entry(root)
name_entry.pack(pady=5)

tk.Label(root, text="Department:").pack(pady=5)
dept_entry = tk.Entry(root)
dept_entry.pack(pady=5)

tk.Button(root, text="Check Employee", command=check_employee).pack(pady=10)

result_label = tk.Label(root, text="", wraplength=350, justify="left")
result_label.pack(pady=10)

root.mainloop()
```
