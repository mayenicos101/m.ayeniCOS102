```python
print("1. Simple Interest")
print("2. Compound Interest")
print("3. Annuity")
print("4. Exit")

while True:
    try:
        choice = int(input("\nEnter your choice (1, 2, 3, or 4): "))
    except ValueError:
        print("Invalid input. Please enter a number between 1 and 4 inclusive.")
        continue
    
    if choice == 1:
        try:
            P = float(input("Enter the principal amount: "))
            R = float(input("Enter the annual interest rate in decimal: "))
            T = float(input("Enter the time in years: "))
            SI = P * R * T
            print("The Simple Interest after", T, "years is:", SI)
        except ValueError:
            print("Invalid input. Please enter numeric values.")
    elif choice == 2:
        try:
            P = float(input("Enter the principal amount: "))
            R = float(input("Enter the annual interest rate in decimal: "))
            T = float(input("Enter the time in years: "))
            N = int(input("Enter the compounding frequency: "))
            A = P * (1 + (R / N)) ** (N * T)
            CI = A - P
            print("The Compound Interest after", T, "years is:", CI)
        except ValueError:
            print("Invalid input. Please enter numeric values.")
        except ZeroDivisionError:
            print("Compounding frequency cannot be zero.")
    elif choice == 3:
        try:
            P = float(input("Enter the principal amount: "))
            R = float(input("Enter the annual interest rate in decimal: "))
            T = float(input("Enter the time in years: "))
            N = int(input("Enter the compounding frequency: "))
            if R == 0:
                A = P * (N * T)
            else:
                A = P * (((1 + (R / N)) ** (N * T) - 1) / (R / N))
            print("The Annuity after", T, "years is:", A)
        except ValueError:
            print("Invalid input. Please enter numeric values.")
        except ZeroDivisionError:
            print("Interest rate cannot be zero for this calculation.")
    elif choice == 4:
        print("Exiting the program...")
        break
    else:
        print("Please enter 1, 2, 3 or 4")
```
